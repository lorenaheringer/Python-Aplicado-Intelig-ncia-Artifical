```python
def escreve_texto_corretamente(texto):
  texto_novo = texto.upper()
  lista_de_palavras = texto_novo.split()
  nome_final = " ".join(lista_de_palavras)
  return nome_final
```

```python
def resumo_emails(lista_emails):
  lista_resumos = []

  for numero, email in enumerate(lista_emails):
    resposta = client.models.generate_content(
        model = "gemini-2.5-flash",
        contents = f"""Estou te enviando o corpo de um email. Quero qur você o resuma em apenas 1 linha,
        passando o intuito daquele email. Segue o email: {email}"""
        )
    print(f"E-mail {numero + 1}: {resposta.text}")
    lista_resumos.append(f"E-mail {numero + 1}: {resposta.text}")

  return lista_resumos
```

```python
lista_resumos = resumo_emails(email_bodies)
```

Desafio:
1. Crie um arquivo .txt com uma lista de perguntas vindas em Python.
2. Ler as perguntas desse arquivo .txt.
3. Obter respostas de um LLm para cada uma.

```python
# Etapa 1
lista_de_perguntas = [
    "De que cor é a laranja?",
    "Qual é a capital do Brasil?",
    "Quem pintou a Mona Lisa?",
    "Onde está localizado o Google?",
    "Qual é a fórmula química da agua?"
]

with open ("lista-de-perguntas.txt", "w", encoding="utf-8") as arquivo:
  for pergunta in lista_de_perguntas:
    arquivo.write(pergunta + "\n")

```

```python
# Etapa 2:
lista_desafio = []

with open ("lista-de-perguntas.txt", "r", encoding="utf-8") as arquivo:
  for linha in arquivo:
    lista_desafio.append(linha.strip())
```

```python
print(lista_desafio)
```

```python
#Etapa 3:
lista_respostas = []

for pergunta in lista_desafio:
  resposta = client.models.generate_content(
      model = "gemini-2.5-flash",
      contents = f"""Gere uma resposta muito sucinta para a pergunta: {pergunta}"""
      )
  lista_respostas.append({"pergunta": pergunta, "resposta": resposta.text})
```

```python
lista_respostas
```

```python
import csv

# Etapa 4:

with open("respostas1.csv", "w", encoding="utf-8", newline='') as arquivo:
  writer = csv.writer(arquivo)
  writer.writerow(["pergunta", "resposta"])
  for pergunta_dict in lista_respostas:
    writer.writerow([pergunta_dict['pergunta'], pergunta_dict['resposta']])
```

```python
#Etapa 2:

lista_de_analises_de_sentimentos = []

for review_numero, resenha in enumerate(coluna_de_reviews):
  resposta = client.models.generate_content(
      model = "gemini-2.5-flash",
      contents = f"""Voce irá analisar a resenha que eu te mandarei abaixo, e retornar
      com uma análise de sentimento. Voce deve responder apenas com uma das seguintes
      palavras: 'Positivo', 'Negativo' ou 'Neutro', indicando o sentimento da resenha.

      Exemplo:
      'Eu adorei esse produto' -> 'Positivo'
      'Esse produto é horrível' -> 'Negativo'
      'O preço é justo', 'satisfaz o proposto' -> 'Neutro'


      Resenha: {resenha}"""
      )
  lista_de_analises_de_sentimentos.append(resposta.text)
  print(f"Resenha {review_numero}: '{resenha}' -> Sentimento: {resposta.text}")
```
